```python
# Importamos las librerias Pandas para el manejo de DataFrames y NumPy para el manejo de arreglos
import numpy as np  # Colocamos un alias a numpy como np
import pandas as pd  # Colocamos un alias a pandas como pd
```

# Introduccion al uso de Pandas
Pandas es una liberia que nos permite manejar dataframes o estrcturas de datos en dos dimensiones que pueden tener varios tipos de datos, las cuales nos permite que sean heterogeneas, similar a una tabla. En esta ocasion podemos tener multiples operaciones con dataframes como ser:

  - Agrupaciones
  - Filtros
  - Uniones
  - Trasnformaciones
  - Lectura de ficheros
  
## Datos utilizaremos para nuestro ejemplo de analisis
Utilizaremos los datos de la declaracion historia de prestamos del banco internacional RF, la cual puedes encontra aqui sobre que es: https://www.bancomundial.org/es/who-we-are/ibrd

En este caso te dejare un enlace para que puedas descargarte ese archivo

```python
# Cargamos la data con la libreria pandas, la cual es un archivo csv 
prestamos_bancarios = pd.read_csv("Declaración histórica de préstamos del BIRF.csv", low_memory=False, sep=",")
```

Si queremos ver la informacion que contiene el archivo puedes hacerlo con el metodo head y luego le indicas cuantos registros quieres que te muestre, por ejemplo 10

```python
# Con el metodo head cargamos solo una cierta cantidad de datos
prestamos_bancarios.head(10)
```

Podemos ordenar nuestra base de datos cargada

```python
# Con el metodo sample ordenamos
prestamos_bancarios = prestamos_bancarios.sample(frac=1)
```

Podemos acceder a la columna de nuestro dataframe

```python
# Con la propiedad column podemos acceder a las columnas de nuestro dataframe
prestamos_bancarios.columns
```

```python
# Con la propiedad dtypes accedemos a los tipos de variables
prestamos_bancarios.dtypes
```

```python
# Con esta accedemos a la propiedad seleccionada dentro de ""
prestamos_bancarios["Region"]
```

```python
# Similar al metodo head
prestamos_bancarios[0:10]
```

```python
# Operacion booleana por propiedad, la cual nos servira para hacer filtrado de datos
prestamos_bancarios["Repaid 3rd Party"] < 100
```

```python
# Seleccion los primeros 10 prestamos bancarios que cumplen la condicion booleana explicada anteriormente
prestamos_bancarios[prestamos_bancarios["Repaid 3rd Party"] < 100].head(10)
```

```python
# Seleccion los primeros 10 prestamos bancarios que cumplen con multiples condiciones como los prestamos bancarios que sean de Honduras y la explicada anteriormente
prestamos_bancarios[(prestamos_bancarios["Repaid 3rd Party"] < 100) & (prestamos_bancarios["Country Code"] == "HN")].head(10)
```

```python
# Utilizando el operador isIn para el filtrado, en donde se seleccionan aquellos registros que cumplan en la propiedad
# Country = "Honduras", "Colombia"
prestamos_bancarios[prestamos_bancarios.Country.isin(["Colombia", "Honduras"])]
```

```python
# Utilizando la libreria pandas filtramos aquellos prestamos bancarios en donde la propiedad "Last Disbursement Date" se encuentra como Nan
prestamos_bancarios[pd.isna(prestamos_bancarios["Last Disbursement Date"])].head(10)
```

```python
# Utilizando la libreria pandas y la funcion len() contamos aquellos prestamos bancarios en donde la propiedad "Last Disbursement Date" se encuentra como Nan
len(prestamos_bancarios[pd.isna(prestamos_bancarios["Last Disbursement Date"])])
```

```python
# Creando la propiedad nombre del pais como Country_Upper
prestamos_bancarios["Country_Upper"] = prestamos_bancarios["Country"].str.upper()
```

```python
# Mostramos la transformacion usando un select
prestamos_bancarios["Country_Upper"]
```

```python
# Eliminamos la columna que hemos creado
del(prestamos_bancarios["Country_Upper"])
```

```python
# Con este borramos las propiedades que colocamos dentro del array e indicamos el eje
prestamos_bancarios.drop(["First Repayment Date"], axis=1)
```

```python
# Con esto borramos los registros de 10 a 80
prestamos_bancarios = prestamos_bancarios.drop(range(10, 80))
```

```python
prestamos_bancarios
```

```python
# Con este agregamos nuevas registros a nuestro dataframe, en donde se agregan con el metodo append()
prestamos_bancarios.append(prestamos_bancarios[prestamos_bancarios["Country Code"] == "HN"])
```

```python
# Agrupar por el codigo de pais y luego sumar la propiedad "Loans Held"
prestamos_bancarios.groupby(by = "Country Code")["Loans Held"].sum()
```

```python
# Agrupar por el codigo de pais y luego sacar la media de la propiedad "Loans Held"
prestamos_bancarios.groupby(by = "Country Code")["Loans Held"].mean()
```

```python
# Agrupar por el codigo de pais y luego describir de la propiedad "Loans Held"
prestamos_bancarios.groupby(by = "Country Code")["Loans Held"].describe()
```

```python
# Agrupar por el codigo de pais y luego sacar el rango de la propiedad "Loans Held"
prestamos_bancarios.groupby(by = "Country Code")["Loans Held"].max() - prestamos_bancarios.groupby(by = "Country Code")["Loans Held"].min()
```

```python
# Agrupar por el "Country Code", "Region" y luego describir de la propiedad 
prestamos_bancarios.groupby(by = ["Country Code", "Region"])["Loans Held"].describe()
```

```python
# Duplicamos los datos del dataframe
prestamos_bancarios.append(prestamos_bancarios)
```

```python
# Con este metodo podemos eliminar los datos duplicados el cual tiene una propiedad llamada subset para eliminar duplicados por columna
prestamos_bancarios.drop_duplicates()
```

```python
# Para elimnar los registros que contienen columna con valor NaN
prestamos_bancarios.dropna(thresh=len(prestamos_bancarios.columns)-1)
```

```python
# Borrar todoas aquellas columnas que en la propiedad "Country Code" este a NaN
prestamos_bancarios.dropna(subset=["Country Code"])
```
